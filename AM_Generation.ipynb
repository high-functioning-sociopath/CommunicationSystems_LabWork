```python
# import required packages
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Amplitude Modulation

# baseband signal:    m(t) = Am*cos(wm*t)
# carrier signal:     c(t) = Ac*cos(wc*t)
# AM signal:          φAM(t) = [Ac + m(t)] * cos(wc*t)
# AM signal:          φAM(t) = Ac[1 + u*cos(wm*t)] * cos(wc*t)
```

```python
#-----------------------------------------------------------------------------------#
#--------------------------- Only for m(t) = Am*cos(wm*t) --------------------------#
#-----------------------------------------------------------------------------------#

print("Note:")
print("     Amplitude unit: Volt (V) (peak-to-peak)")
print("     Frequency unit: Hertz (Hz)")
print()
A_m = float(input("Enter Baseband Amplitude: ")) #10
f_m = float(input("Enter Baseband Frequency: ")) #10
A_c = float(input("Enter Carrier Amplitude: ")) #20
f_c = float(input("Enter Carrier Frequency: ")) #60
u = float(input("Enter Modulation Index: "))#1

# signal generation

t = np.linspace(0, 1, 10000)

# w_m = 2*pi*f_m, similar for w_c
baseband = A_m*np.cos(2*np.pi*f_m*t)
carrier = A_c*np.cos(2*np.pi*f_c*t)
amplitude_modulated = baseband*carrier
am_wave_using_mod_index = A_c * (1 + (u*np.cos(2*np.pi*f_m*t))) * np.cos(2*np.pi*f_c*t)

# plotting signals

plt.subplot(4,1,1)
plt.title("Baseband signal (m(t))")
plt.plot(baseband,'g')
plt.xlabel("Time")

plt.subplot(4,1,2)
plt.title("Carrier signal (c(t))")
plt.plot(carrier, 'r')
plt.xlabel("Time")

plt.subplot(4,1,3)
plt.title("AM signal (φDSB-SC(t))")
plt.plot(amplitude_modulated, color="blue")
plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.subplot(4,1,4)
plt.title("AM signal (φAM(t))")
plt.plot(am_wave_using_mod_index, color="magenta")
plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.suptitle("Amplitude Modulation")
fig.set_size_inches(16, 11)
```

```python
#-----------------------------------------------------------------------------------#
#--------------------------- For any input m(t) --------------------------#
#-----------------------------------------------------------------------------------#

print("Note: ")
print("     1. Frequency in Hertz (Hz)")
print("     2. baseband signal, m(t) should be of proper format")
print("           Eg: 10*np.cos(2*np.pi*5*t)", end = "\n")
print("           Eg: 30*signal.sawtooth(2*np.pi*10*t)   <- for ramp", end = "\n")
print("           Eg: 30*signal.sawtooth(2*np.pi*10*t,0.5)   <- for triangle", end = "\n")
print("           Eg: 1*signal.square(2 * np.pi * 5 * t)   <- for square", end = "\n")

print("Check documents for more details")
print()

# inputs
m_t_string = str(input("Enter baseband signal m(t): "))
f_c = float(input("Enter Carrier Frequency: "))
u = float(input("Enter Modulation Index: "))

print()

t = np.linspace(0, 1, 500)

baseband = eval(m_t_string) #baseband signal

# mod_index = max m(t) / A_c
a_c = (int(np.max(baseband)) // u) or 10 #carrier amplitude

carrier = a_c*np.cos(2*np.pi*f_c*t)
amplitude_modulated = (a_c + baseband) * np.cos(2 * np.pi * f_c * t)

plt.subplot(3,1,1)
plt.title("Baseband signal (m(t))")
plt.plot(baseband,'green')
plt.xlabel("Time")

plt.subplot(3,1,2)
plt.title("Carrier signal (c(t))")
plt.plot(carrier, 'red')
plt.xlabel("Time")

plt.subplot(3,1,3)
plt.title("AM signal (φAM(t))")
plt.plot(amplitude_modulated, color="blue")
plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.suptitle("Amplitude Modulation")
fig.set_size_inches(16, 11)
```
